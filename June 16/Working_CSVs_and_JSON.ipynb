```python
from pyspark.sql import functions as F
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("SalesData") \
    .master("local[*]") \
    .getOrCreate()
```

```python
flights_csv_content = """booking_id,passenger_name,from_city,to_city,airline,ticket_price,travel_class,status
1001,Aarav Mehta,Hyderabad,Delhi,IndiGo,6500,Economy,Confirmed
1002,Sana Khan,Bangalore,Mumbai,Vistara,8200,Economy,Confirmed
1003,John Mathew,Chennai,Delhi,Air India,12000,Business,Confirmed
1004,Ayesha Begum,Hyderabad,Dubai,Emirates,28000,Economy,Confirmed
1005,Vikram Rao,Mumbai,Singapore,Singapore Airlines,35000,Business,Pending
1006,Divya Sharma,Delhi,Hyderabad,IndiGo,5900,Economy,Cancelled
1007,Imran Ali,Pune,Bangalore,Akasa Air,4800,Economy,Confirmed
1008,Meera Nair,Kochi,Dubai,Emirates,26000,Economy,Confirmed
1009,Rohan Das,Kolkata,Delhi,Air India,7400,Economy,Pending
1010,Nisha Reddy,Bangalore,London,British Airways,62000,Business,Confirmed
"""
with open("flights.csv", "w") as f:
    f.write(flights_csv_content)
```

```python
flights_df = spark.read.csv(
    "flights.csv",
    header=True,
    inferSchema=True,
)
flights_df.show()
flights_df.printSchema()
```

```python
flights_df.show()
```

```python
flights_df.select("passenger_name", "from_city", "to_city").show()
```

```python
flights_df.filter(flights_df.status == "Confirmed").show()
```

```python
flights_df.filter(flights_df.status == "Cancelled").show()
```

```python
flights_df.filter(flights_df.status == "Pending").show()
```

```python
flights_df.filter(flights_df.ticket_price > 20000).show()
```

```python
flights_df.filter(
    flights_df.to_city.isin("Dubai", "Singapore", "London")
).show()
```

```python
print(flights_df.count())
```

```python
flights_df.groupBy("airline").count().show()
```

```python
flights_df.groupBy("status").count().show()
```

```python
flights_df.filter(flights_df.status == "Confirmed") \
    .agg(F.sum("ticket_price").alias("total_revenue")).show()
```

```python
flights_df.groupBy("airline").agg(
    F.avg("ticket_price").alias("avg_ticket_price")
).show()
```

```python
flights_df.agg(F.max("ticket_price").alias("highest_price")).show()
```

```python
flights_df.agg(F.min("ticket_price").alias("lowest_price")).show()
```

```python
flights_df.orderBy(F.desc("ticket_price")).show()
```

```python
flights_df2 = flights_df.withColumn("tax", F.col("ticket_price") * 0.05)
flights_df2.show()
```

```python
flights_df2 = flights_df2.withColumn(
    "final_price", F.col("ticket_price") + F.col("tax")
)
flights_df2.show()
```

```python
flights_df2 = flights_df2.withColumn(
    "price_category",
    F.when(F.col("ticket_price") >= 30000, "Premium")
     .when(F.col("ticket_price") >= 10000, "Standard")
     .otherwise("Budget"),
)
flights_df2.show()
```

```python
flights_df2.groupBy("price_category").count().show()
```

```python
flights_df.filter(flights_df.status == "Confirmed") \
    .write.mode("overwrite").parquet("confirmed_flights.parquet")
```

```python
hotels_df = spark.read.option("multiline", "true").json("hotels.json")
hotels_df.show(truncate=False)
hotels_df.printSchema()
```

```python
hotels_df2 = hotels_df.withColumn(
    "total_potential_revenue",
    F.col("rooms_available") * F.col("price_per_night"),
)
hotels_df2.show(truncate=False)
```

```python
hotels_flattened = hotels_df.select(
    "hotel_id",
    "hotel_name",
    "city",
    "category",
    "rating",
    "rooms_available",
    "price_per_night",
    F.col("contact.phone").alias("contact_phone"),
    F.col("contact.email").alias("contact_email"),
    F.concat_ws(",", F.col("amenities")).alias("amenities"),
)
hotels_flattened.show(truncate=False)
hotels_flattened.write.mode("overwrite").parquet("hotels_flattened.parquet")
```
