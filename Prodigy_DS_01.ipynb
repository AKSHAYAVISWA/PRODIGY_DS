```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "C:\\Users\\admin\\Desktop\\API_SP.POP.TOTL_DS2_en_csv_v2_9949.csv" # Change this to your file path
data = pd.read_csv(file_path,skiprows=4,sep=',')
```

```python
data.head()
```

```python
data_2023 = data[data['Indicator Code'] == 'SP.POP.TOTL']
```

```python
top_10_most_populated = data_2023.nlargest(10, '2023')
```

```python
least_populated = data_2023.nsmallest(10, '2023')
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_10_most_populated['Country Name'], top_10_most_populated['2023'])
plt.title('Top 10 Most Populated Countries(2023)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=60,ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(least_populated['Country Name'], least_populated['2023'])
plt.title('10 Least Populated Countries(2023)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=60,ha='right')
plt.tight_layout()
plt.show()
```
