```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("C:\\Users\\admin\\Desktop\\dataset\\RTA Dataset.csv")
```

```python
data['Time'] = pd.to_datetime(data['Time'], format='%H:%M:%S')
data['Day_of_week'] = pd.Categorical(data['Day_of_week'], categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)
```

```python
data.fillna({'Cause_of_accident': 'Unknown', 'Casualty_class': 'na'}, inplace=True)
```

```python
le = LabelEncoder()
data['Sex_of_driver_encoded'] = le.fit_transform(data['Sex_of_driver'])
data['Type_of_vehicle_encoded'] = le.fit_transform(data['Type_of_vehicle'])
data['Accident_severity_encoded'] = le.fit_transform(data['Accident_severity'])
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=data['Time'].dt.hour, palette='coolwarm')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_week', data=data, palette='Set2')
plt.title('Accidents by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_band_of_driver', hue='Accident_severity', data=data, palette='viridis')
plt.title('Accident Severity by Driver Age')
plt.xlabel('Driver Age Band')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
vehicle_counts = data['Type_of_vehicle'].value_counts().head(10)  # Top 10 vehicle types
plt.figure(figsize=(10, 6))
sns.countplot(x='Type_of_vehicle', data=data[data['Type_of_vehicle'].isin(vehicle_counts.index)], palette='Set2')
plt.title('Top 10 Accidents by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Driving_experience', hue='Accident_severity', data=data, palette='Blues')
plt.title('Accident Severity by Driving Experience')
plt.xlabel('Driving Experience')
plt.ylabel('Number of Accidents')
plt.show()
```
